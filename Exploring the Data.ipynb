# Exploring the Data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('./train.csv', header=0)
```

## Family Size and Class
Checking to see if there is a pattern between family size and class.

```python
data['fam_size'] = data['SibSp'] + data['Parch'] + 1
```

```python
fam_class = []
for i in data['Pclass'].unique():
    fam_class.append(data[data['Pclass'] == i]['fam_size'].tolist())
plt.boxplot(fam_class)
plt.show()
```

# Age vs Fare

```python
plt.plot(data[data['Survived']==0]['Age'], data[data['Survived']==0]['Fare'], '.r',
         data[data['Survived']==1]['Age'], data[data['Survived']==1]['Fare'], '.b')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
